# Assignment 4: Data Visualization (Python)

### Load the Necessary Libraries

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Datasets (2019 to 2023)

```python
# Load the datasets
file_path_2023 = '../../../02_activities/assignments/assignment_4/data/en-2023-pssd.csv'
data_2023 = pd.read_csv(file_path_2023)

file_path_2022 = '../../../02_activities/assignments/assignment_4/data/en-2022-pssd.csv'
data_2022 = pd.read_csv(file_path_2022)

file_path_2021 = '../../../02_activities/assignments/assignment_4/data/en-2021-pssd.csv'
data_2021 = pd.read_csv(file_path_2021)

file_path_2020 = '../../../02_activities/assignments/assignment_4/data/en-2020-pssd.csv'
data_2020 = pd.read_csv(file_path_2020)

file_path_2019 = '../../../02_activities/assignments/assignment_4/data/en-2019-pssd.csv'
data_2019 = pd.read_csv(file_path_2019)
```

## Schema
The dataset contains the following columns:

| column | description |
|--------|-------------|
| Sector | The sector in which the individual is employed. |
| Last Name | The last name of the individual. |
| First Name | The first name of the individual. |
| Salary | The salary of the individual. |
| Benefits | The benefits received by the individual. |
| Employer | The employer of the individual. |
| JobTitle | The job title of the individual. |
| Year | The year of the data entry. |

## Data Cleaning (2023)

```python
# Filter the data to include only relevant columns
data_filtered_1 = data_2023[['Sector', 'Salary']].dropna()

# Ensure the Salary column is numeric
data_filtered_1['Salary'] = pd.to_numeric(data_filtered_1['Salary'], errors='coerce')

# Drop any rows with NaN values in 'Salary' after conversion
data_filtered_1 = data_filtered_1.dropna(subset=['Salary'])

# Calculate the average salary by sector
avg_salary_by_sector_1 = data_filtered_1.groupby('Sector')['Salary'].mean().reset_index()
```

## Data Cleaning (2019 - 2022)

```python
# Repeat the data manipulation process for 2019 to 2022

# Data manipulation for 2022
data_filtered_2 = data_2022[['Sector', 'Salary paid']].dropna()
data_filtered_2['Salary paid'] = pd.to_numeric(data_filtered_2['Salary paid'], errors='coerce')
data_filtered_2 = data_filtered_2.dropna(subset=['Salary paid'])
avg_salary_by_sector_2 = data_filtered_2.groupby('Sector')['Salary paid'].mean().reset_index()

# Data manipulation for 2021
data_filtered_3 = data_2021[['Sector', 'Salary']].dropna()
data_filtered_3['Salary'] = pd.to_numeric(data_filtered_3['Salary'], errors='coerce')
data_filtered_3 = data_filtered_3.dropna(subset=['Salary'])
avg_salary_by_sector_3 = data_filtered_3.groupby('Sector')['Salary'].mean().reset_index()

# Data manipulation for 2020
data_2020['Salary'] = data_2020['Salary'].replace('[\$,]', '', regex=True) # DEBUGGING: Remove dollar signs and commas from the 'Salary' column
data_2020['Salary'] = pd.to_numeric(data_2020['Salary'], errors='coerce')
data_filtered_4 = data_2020[['Sector', 'Salary']].dropna()
avg_salary_by_sector_4 = data_filtered_4.groupby('Sector')['Salary'].mean().reset_index()

# Data manipulation for 2019
data_filtered_5 = data_2019[['Sector', 'Salary Paid']].dropna()
data_filtered_5['Salary Paid'] = pd.to_numeric(data_filtered_5['Salary Paid'], errors='coerce')
data_filtered_5 = data_filtered_5.dropna(subset=['Salary Paid'])
avg_salary_by_sector_5 = data_filtered_5.groupby('Sector')['Salary Paid'].mean().reset_index()
```

## Plotting the Average Salary by Sector (2023)

```python
# Create the bar chart for average salary by sector
plt.figure(figsize=(14, 8))
sns.barplot(x='Salary',
            y='Sector',
            data=avg_salary_by_sector_1,
            hue='Sector',
            palette='viridis',
            dodge=False,
            legend=False)

# creating two functions to define characteristics for Fonts
font1 = {'family':'sans-serif','size':22}
font2 = {'family':'sans-serif','size':18}

# setting the plot titles, labels and fonts
plt.title('Average Salary by Sector (2023)', fontdict=font1)
plt.xlabel('Average Salary', fontdict=font2)
plt.ylabel('Sector', fontdict=font2)

# running the plot
plt.show()
```
