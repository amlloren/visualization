```python
np.random.seed(613)
x = np.arange(50) 
y = np.random.randint(0,100,50) # 1st value is our min, 2nd is max, 3rd is how many to plot

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

### Adding Labels

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

### Modifying labels and titles

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total Growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

### Modifying data points

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker = "*",
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'indigo',
           linestyle = '--',
           linewidth = 1)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'indigo',
           linestyle = '--',
           linewidth = 1,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
```

### Grid lines

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'pink',
           linestyle = '--',
           linewidth = 1,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid()
```

### MAKE YOUR OWN

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = ".",
        color = 'blue',
        linestyle = ':',
        linewidth = 1,
        markeredgecolor = '#FF5F5f',
        markerfacecolor = '#000000')

ax.grid()
```

## Activity

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})
```

```python
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.5 # transparency
         ) 

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```
