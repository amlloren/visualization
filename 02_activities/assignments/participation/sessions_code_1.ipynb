```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import seaborn as sns
```

# Session 1: November 19, 2024

```python
# Creating a Scatter Plot to explore daata visualization methods.

np.random.seed(613) # Setting a seed for reproducibility
x = np.arange(50)  # Setting the range to 0 to 49.
y = np.random.randint(0,100,50) # 1st value is our min, 2nd is max, 3rd is how many to plot

fig, ax = plt.subplots(figsize=(5,3)) # Setting the dimensions to 5 by 3
ax.scatter(x,y)
```

```python
# Exploring another way to visualize our data as a Bar Graph
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# Exploring another way to visualize our data as a Horizontal Bar Graph
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y) # Creates horizontal bar graphs
```

```python
# Exploring another way to visualize our data as a Plot Line Graph

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

### Adding Labels

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time') # Adding a Title to the graph
ax.set_ylabel('Total Growth') # Adding a label to the y axis
ax.set_xlabel('Years since start') # Adding a label to the x axis
```

### Modifying labels and titles

```python
# creating two functions to define characteristics for Fonts
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1) # Adding a Title and setting a font style
ax.set_ylabel('Total Growth', fontdict=font2) # Adding a label on my y axis and setting a font style
ax.set_xlabel('Years since start', fontdict=font2) # Adding a label on my x axis and setting a font style
```

### Modifying data points

```python
# Exploring how to modify data points
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker = "*", # Setting the marker to a star
           color = 'indigo') # setting the color to indigo
```

```python
# Exploring how to modify our data visualization
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'indigo',
           linestyle = '--', # Added a dashed line
           linewidth = 1)
```

```python
# Exploring how to modify our data visualization
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'indigo',
           linestyle = '--',
           linewidth = 1,
           markeredgecolor = '#FA9359', # This sets a color to the ourline of the marker
           markerfacecolor = '#000000') # This sets the color of the inside of the marker
```

### Grid lines

```python
# Exploring how to modify the grid lines in a data visualization
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = "*",
           color = 'pink',
           linestyle = '--',
           linewidth = 1,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid() #Adds a grid to the graph
```

### MAKE YOUR OWN

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = ".", # Changed the marker to circles
        color = 'blue',
        linestyle = ':', # Set dotted lines
        linewidth = 1,
        markeredgecolor = '#FF5F5f',
        markerfacecolor = '#000000')

ax.grid()
```

```python
# Creating Data Frame to be used
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})
```

```python
# Import necessary libraries and packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

### Activity: Try to plot both y variables on the same axis

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1") # Plotting Y1 and add label
ax.plot(x,y2, label = "Person 2") # Plotting Y2 and add label

# Customizing the labelling for this graph
ax.legend(loc="lower right",
          frameon = True,
          fontsize=12,
          ncol=2,
          shadow=True) 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1") # Plotting Y1 and add label
ax.plot(x,y2, label = "Person 2") # Plotting Y2 and add label

# Customizing the labelling placement for this graph
ax.legend(loc="upper left",
          bbox_to_anchor=(1,1)) # Setting the pinning of the legend
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label='Person 1')
ax.scatter(x,y2,label='Person 2')
ax.legend(loc="upper left",
          bbox_to_anchor=(1,1))

# Modifying the overlaying text
ax.text(10,95,"This value is important!",
        ha="center",
        color = 'red',
        size = 20)
```

#### Positioning Text and Annotations

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1,5,". Data:(1,5)", transform=ax.transData) # In reference to the Data
ax.text(0.5,0.1,". Axes: (0.5, 0.1)", transform=ax.transAxes) # In referencee to the Axes
ax.text(0.2,0.2,". Figure: (0.2,0.2)", transform=fig.transFigure) # In reference to the Figure
```

### Annotating with arrows

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")

ax.legend(loc='lower right')

ax.annotate('This is important!'
, xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black'
, shrink=0.05))

fig.show()
```

### Modifying arrows

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")

ax.legend(loc='lower right')

ax.annotate("This is important!", xy=(10, 95), xytext=(20, 94), arrowprops = dict(arrowstyle = "wedge", color = " hotpink "))

fig.show()
```
