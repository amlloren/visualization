```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji","Chopper","Robin"])
y2 = np.array([110, 180, 240, 99, 220, 140, 210])
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3))

color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.5 # transparency
         ) 

ax1.scatter(x1,y1)
ax2.bar(x2,y2,color=color)

fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))

color = (0.2,  # redness
         0.4,  # greenness
         0.2,  # blueness
         0.5)  # transparency

# Scatter plot on ax1
ax1.scatter(x1, y1)
# Rotate x-axis tick labels
ax1.tick_params(axis='x', rotation=45)
ax1.set_title('Scatter Plot')

# Bar plot on ax2
ax2.bar(x2, y2, color=color)
# Rotate x-axis tick labels
ax2.tick_params(axis='x', rotation=45)
ax2.set_title('Bar Plot')

ax2.text(4.5,140,"MA FAVE", transform=ax2.transData) # In reference to the Data

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(11, 5))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(9, 4),
                                    layout = "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2, color = 'green')
someaxes["ax3"].plot(x1,y1, color = 'violet')
someaxes["ax1"].set_xlabel('A Big Label',fontsize=16)
someaxes["ax2"].set_xlabel('One Piece Bounty',fontsize=16)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=16)

plt.show()
```
