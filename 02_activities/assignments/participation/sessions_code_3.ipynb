```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
from PIL import Image
import requests
from io import BytesIO
```

```python
# first make our sample data
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = "none")

# The error bars are showing the standard deviation
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")

# Modifying visualization of error bars
ax.errorbar(x,
            y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")

# Modifying visualization of error bars
ax.errorbar(x,
            y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4,
            errorevery=2)
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red", linestyle = ':')

ax_image = fig.add_axes([0.08, 0.5, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis("off")
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid') # Setting style for the plot

sns.lineplot(data=tips, # choose our dataset
             x='total_bill', # define our x variable
             y='tip') # define our y variable

# Set_style - whitegrid, white, darkgrid
```

```python
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
tipgraph = sns.scatterplot(data=tips, x='total_bill',
                           y='tip', 
                           style = 'time', 
                           hue ='day',
                           palette = ['purple', 'hotpink','deepskyblue','yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
sns.pairplot(data = tips, hue = 'day')
```

```python
daysplot = sns.relplot(data=tips,
                       x="total_bill", # Set the x axis to 'total_bill'
                       y="tip", # Set y axis to 'tip'
                       hue="sex", # set colors of points based on 'sex'
                       col="day", # Set separate plots for each day
                       kind="scatter",# Set kind of plot as a Scatter plot
                       col_wrap=2 # Seting two plots per line
                       )
```

## Making wordclouds

```python
from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))

ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
          interpolation='bilinear') # this line helps smooth our image

ax.axis("off")
plt.show()
```

## Making venn diagram

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
A = set(["apple","banana","watermelon"])
B = set(["pumpkin","blueberry","apple","key lime"])
```

```python
diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits','Pies'),
                           set_colors=("blue","red"),
                           alpha=0.5)

plt.show()
```

```python
diagram = venn2([A, B],
                           set_labels = ('Fruits','Pies'),
                           set_colors=("blue","red"),
                           alpha=0.5)

plt.show()
```

```python
diagram = venn2([A, B],
                           set_labels = ('Fruits','Pies'),
                           set_colors=("blue","red"),
                           alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))

plt.show()
```
